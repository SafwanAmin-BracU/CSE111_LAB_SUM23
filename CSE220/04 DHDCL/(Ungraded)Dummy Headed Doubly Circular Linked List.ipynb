```python
class Node:
    def __init__(self, e, n, p):
        self.element = e
        self.next = n
        self.prev = p
```

```python
class DoublyList:
    # Creates a Dummy Head for Dummy Headed Circular Doubly Circular Linked List.
    def __init__(self, elem=None, next=None, prev=None):
        self.dh: Node = Node(elem, next, prev)
        self.dh.next = self.dh
        self.dh.prev = self.dh

    # create a DHDCL from an array
    def create_list(self, l):
        temp = self.dh
        for i in range(len(l)):
            newNode = Node(l[i], None, temp)
            temp.next = newNode  # type: ignore
            temp = temp.next

        temp.next = self.dh  # type: ignore
        self.dh.prev = temp

    # Counts the number of Nodes in the list and return the number
    def countNode(self):
        temp = self.dh.next
        count = 0
        while temp != self.dh:
            count += 1
            temp = temp.next
        return count

    # prints the elements in the list forward
    def forwardprint(self):
        temp = self.dh.next
        while temp != self.dh:
            endStr = "," if temp.next != self.dh else ".\n"
            print(temp.element, end=endStr)
            temp = temp.next

    # prints the elements in the list backward
    def backwardprint(self):
        temp = self.dh.prev
        while temp != self.dh:
            endStr = "," if temp.prev != self.dh else ".\n"
            print(temp.element, end=endStr)
            temp = temp.prev


    # returns the reference of the at the given index. For invalid index return None.
    def nodeAt(self, idx):
        if idx < 0 or idx >= self.countNode():
            return None
        temp = self.dh
        for i in range(idx + 1):
            temp = temp.next
        return temp


    def indexOf(self, elem):
        temp = self.dh.next
        idx = 0
        while temp != self.dh:
            if temp.element == elem:
                return idx
            temp = temp.next
            idx += 1
        return -1


    # insert at specific index of the DHDCL
    def insert(self, elem, idx):
        if idx < 0 or idx > self.countNode():
            return None
        temp = self.dh
        for i in range(idx):
            temp = temp.next
        newNode = Node(elem, temp.next, temp)
        temp.next = newNode
        newNode.next.prev = newNode


    # removes the node of the specific index and returns the element of the node
    def remove(self, idx):
        if idx < 0 or idx >= self.countNode():
            return None
        temp = self.dh
        for i in range(idx):
            temp = temp.next
        var = temp.next.element
        temp.next = temp.next.next
        temp.next.prev = temp
        return var
```

```python
import unittest
from unittest.mock import patch
import io


class TestDoublyList(unittest.TestCase):
    def setUp(self) -> None:
        self.h1 = DoublyList()
        self.h2 = DoublyList()

    def test_test01(self):
        expected_output = [
            "Forward Print: 10,20,30,40.\n",
            "Backward Print: 40,30,20,10.\n",
            "Total Nodes: 4\n",
        ]

        with patch("sys.stdout", new_callable=io.StringIO) as mock_stdout:
            self.h1.create_list(np.array([10, 20, 30, 40]))
            print("Forward Print:", end=" ")
            self.h1.forwardprint()
            print("Backward Print:", end=" ")
            self.h1.backwardprint()
            print("Total Nodes:", end=" ")
            print(self.h1.countNode())

            actual_output = mock_stdout.getvalue()

        for i, line in enumerate(actual_output.splitlines()):
            self.assertEqual(line.strip(), expected_output[i].strip())

    def test_test02(self):
        expected_output = [
            "Node at 2nd index: 30\n",
            "Node at 0th index: 10\n",
            "Node at -1th index: Index Error\n",
        ]

        with patch("sys.stdout", new_callable=io.StringIO) as mock_stdout:
            n = self.h1.nodeAt(2)
            print(
                "Node at 2nd index: "
                + str(n.element if n is not None else "Index Error")
            )
            n = self.h1.nodeAt(0)
            print(
                "Node at 0th index: "
                + str(n.element if n is not None else "Index Error")
            )
            n = self.h1.nodeAt(-1)
            print(
                "Node at -1th index: "
                + str(n.element if n is not None else "Index Error")
            )

            actual_output = mock_stdout.getvalue()

        for i, line in enumerate(actual_output.splitlines()):
            self.assertEqual(line.strip(), expected_output[i].strip())

    def test_test03(self):
        expected_output = ["Index of element 40: 3\n",
                           "Index of element 6: -1\n"]

        with patch("sys.stdout", new_callable=io.StringIO) as mock_stdout:
            index = self.h1.indexOf(40)
            print("Index of element 40: " + str(index))
            index = self.h1.indexOf(6)
            print("Index of element 6: " + str(index))

            actual_output = mock_stdout.getvalue()

        for i, line in enumerate(actual_output.splitlines()):
            self.assertEqual(line.strip(), expected_output[i].strip())

    def test_test04(self):
        expected_output = [
            "Inserting 85 in 0th index\n",
            "Forward Print: 85,10,20,30,40.\n",
            "Backward Print: 40,30,20,10,85.\n",
            "Inserting 95 in 3rd index\n",
            "Forward Print: 85,10,20,95,30,40.\n",
            "Backward Print: 40,30,95,20,10,85.\n",
            "Inserting 75 in 6th index\n",
            "Forward Print: 85,10,20,95,30,40,75.\n",
            "Backward Print: 75,40,30,95,20,10,85.\n",
        ]

        with patch("sys.stdout", new_callable=io.StringIO) as mock_stdout:
            self.h1.insert(85, 0)
            print("Inserting 85 in 0th index")
            self.h1.forwardprint()
            self.h1.backwardprint()
            self.h1.insert(95, 3)
            print("Inserting 95 in 3rd index")
            self.h1.forwardprint()
            self.h1.backwardprint()
            self.h1.insert(75, 6)
            print("Inserting 75 in 6th index")
            self.h1.forwardprint()
            self.h1.backwardprint()

            actual_output = mock_stdout.getvalue()

        for i, line in enumerate(actual_output.splitlines()):
            self.assertEqual(line.strip(), expected_output[i].strip())

    def test_test05(self):
        expected_output = [
            "Forward Print: 10,20,30,40,50,60,70.\n",
            "Removed element: 10\n",
            "Forward Print: 20,30,40,50,60,70.\n",
            "Backward Print: 70,60,50,40,30,20.\n",
            "Removed element: 50\n",
            "Forward Print: 20,30,40,60,70.\n",
            "Backward Print: 70,60,40,30,20.\n",
            "Removed element: 70\n",
            "Forward Print: 20,30,40,60.\n",
            "Backward Print: 60,40,30,20.\n",
        ]

        with patch("sys.stdout", new_callable=io.StringIO) as mock_stdout:
            self.h2.create_list(np.array([10, 20, 30, 40, 50, 60, 70]))
            print("Forward Print: ", end=" ")
            self.h2.forwardprint()
            print("Removing node from 0th index")
            print("Removed element: " + str(self.h2.remove(0)))
            print("Forward Print: ", end=" ")
            self.h2.forwardprint()
            print("Backward Print: ", end=" ")
            self.h2.backwardprint()
            print("Removing node from 3rd index")
            print("Removed element: " + str(self.h2.remove(3)))
            print("Forward Print: ", end=" ")
            self.h2.forwardprint()
            print("Backward Print: ", end=" ")
            self.h2.backwardprint()
            print("Removing node from 4th index")
            print("Removed element: " + str(self.h2.remove(4)))
            print("Forward Print: ", end=" ")
            self.h2.forwardprint()
            print("Backward Print: ", end=" ")
            self.h2.backwardprint()

            actual_output = mock_stdout.getvalue()

        for i, line in enumerate(actual_output.splitlines()):
            self.assertEqual(line.strip(), expected_output[i].strip())


# Run the unit tests.
unittest.main(argv=[""], verbosity=2, exit=False)
```

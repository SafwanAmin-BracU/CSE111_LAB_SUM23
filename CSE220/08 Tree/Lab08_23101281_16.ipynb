```python
class BTNode:
  def __init__(self, elem):
    self.elem = elem
    self.right = None
    self.left = None
```

```python
def inorder(root):
  if root == None:
    return

  inorder(root.left)
  print(root.elem, end = ' ')
  inorder(root.right)
```

```python
def tree_construction(arr, i = 1):
  if i>=len(arr) or arr[i] == None:
    return None
  p = BTNode(arr[i])
  p.left = tree_construction(arr, 2*i)
  p.right = tree_construction(arr, 2*i+1)
  return p


root2 = tree_construction([None, 'A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', None, None, None, 'I', 'J', None, 'k'])
inorder(root2)
```

```python
def convert_mirror(root):
    if root.left != None:
        convert_mirror(root.left)
    if root.right != None:
        convert_mirror(root.right)
    root.left, root.right = root.right, root.left
    return root



# DRIVER CODE


root = BTNode(10)
n1 = BTNode(20)
n2 = BTNode(30)
n3 = BTNode(40)
n4 = BTNode(60)


root.left = n1
root.right = n2


n1.left = n3
n1.right = n4



print("Given Tree Inorder Traversal: ", end=" ")
inorder(root)  # Given Tree Inorder Traversal:  40 20 60 10 30
print()


root2 = convert_mirror(root)
print("Mirrored Tree Inorder Traversal: ", end=" ")
inorder(root2)  # Mirrored Tree Inorder Traversal:  30 10 60 20 40
```

```python
def smallest_level(root, nLevel=0, levelsValueMap={}):
    if root == None:
        return levelsValueMap
    if nLevel in levelsValueMap:
        levelsValueMap[nLevel] = min(levelsValueMap[nLevel], root.elem)
    else:
        levelsValueMap[nLevel] = root.elem
    if root.left != None:
        smallest_level(root.left, nLevel + 1, levelsValueMap)
    if root.right != None:
        smallest_level(root.right, nLevel + 1, levelsValueMap)
    return levelsValueMap



# DRIVER CODE


root = tree_construction([None, 4, 9, 2, 3, -5, None, 7])
print("Given Tree Inorder Traversal: ", end=" ")
inorder(root)  # Given Tree Inorder Traversal:  3 9 5 4 2 7
print()
print("Level Wise Smallest Value: ", end=" ")
print(smallest_level(root))  # Level Wise Smallest Value:  {0: 4, 1: 2, 2: -5}
```

```python
def inorder_predecessor(root, x):
    if root == None:
        return None
    if root.elem == x.elem:
        if root.left != None:
            p = root.left
            while p.right != None:
                p = p.right
            return p
    elif x.elem < root.elem:
        return inorder_predecessor(root.left, x)
    else:
        return inorder_predecessor(root.right, x)
        

#DRIVER CODE
root = BTNode(20)
n1 = BTNode(8)
n2 = BTNode(22)
n3 = BTNode(4)
n4 = BTNode(12)
n5 = BTNode(10)
n6 = BTNode(14)

root.left = n1
root.right = n2

n1.left = n3
n1.right = n4

n4.left = n5
n4.right = n6

print('Given Tree Inorder Traversal: ', end = ' ')
inorder(root) #Given Tree Inorder Traversal:  4 8 10 12 14 20 22
print()

x = root
print(f'Inorder predecessor of node {x.elem}: {inorder_predecessor(root, x).elem}') #Inorder predecessor of node 20: 14
```

```python
def LCA(root, x, y):
    if x < root.elem and y < root.elem:
        return LCA(root.left, x, y)
    elif x > root.elem and y > root.elem:
        return LCA(root.right, x, y)
    return root.elem


# DRIVER CODE


# Write by yourself from the given tree
root = BTNode(15)
root.left = BTNode(10)
root.right = BTNode(25)

root.left.left = BTNode(8)
root.left.right = BTNode(12)

root.left.left.left = BTNode(5)
root.left.left.right = BTNode(9)

root.right.left = BTNode(20)
root.right.right = BTNode(30)

root.right.left.left = BTNode(18)
root.right.left.right = BTNode(22)


t1 = LCA(root, 6, 12)
t2 = LCA(root, 20, 6)
t3 = LCA(root, 18, 22)
t4 = LCA(root, 20, 25)
t5 = LCA(root, 10, 12)

print(f'LCA of 6 and 12: {t1}')  # LCA of 6 and 12: 10
print(f'LCA of 20 and 6: {t2}')  # LCA of 20 and 6: 15
print(f'LCA of 18 and 22: {t3}')  # LCA of 18 and 22: 20
print(f'LCA of 20 and 25: {t4}')  # LCA of 20 and 25: 25
print(f'LCA of 10 and 12: {t5}')  # LCA of 10 and 12: 10
```

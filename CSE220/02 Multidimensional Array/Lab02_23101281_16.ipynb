```python
# You must run this cell to print matrix and for the driver code to work
def print_matrix(m: np.ndarray):
    row, col = m.shape
    for i in range(row):
        c = 1
        print("|", end="")
        for j in range(col):
            c += 1
            if len(str(m[i][j])) == 1:
                print(" ", m[i][j], end="  |")
                c += 6
            else:
                print(" ", m[i][j], end=" |")
                c += 6
        print()
        print("-" * (c - col))
```

#Zigzag Walk


```python
def walk_zigzag(floor: np.ndarray) -> None:
    rows, cols = floor.shape
    for c in range(cols):
        for r in range(rows):
            if c % 2 == 0 and r % 2 == 0:
                print(floor[r][c], end=" ")
            if c % 2 == 1 and r % 2 == 1:
                r = rows + ~r
                print(floor[r][c], end=" ")
        print()


floor = np.array(
    [
        ["3", "8", "4", "6", "1"],
        ["7", "2", "1", "9", "3"],
        ["9", "0", "7", "5", "8"],
        ["2", "1", "3", "4", "0"],
        ["1", "4", "2", "8", "6"],
    ]
)

print_matrix(floor)
print("Walking Sequence:")
walk_zigzag(floor)
# This should print
# 3 9 1
# 1 2
# 4 7 2
# 4 9
# 1 8 6
print("################")
floor = np.array(
    [
        ["3", "8", "4", "6", "1"],
        ["7", "2", "1", "9", "3"],
        ["9", "0", "7", "5", "8"],
        ["2", "1", "3", "4", "0"],
    ]
)

print_matrix(floor)
print("Walking Sequence:")
walk_zigzag(floor)
# This should print
# 3 9
# 1 2
# 4 7
# 4 9
# 1 8
```

#Wall Up Trost District


```python
def create_fence(district: np.ndarray, depth: int) -> np.ndarray:
    district_row, district_col = district.shape
    output = np.array([[8] * (district_col + depth * 2)] * (district_row + depth * 2))
    n = len(output)
    for i in range(district_row * district_col):
        r, c = i // district_col, i % district_col
        output[r + depth][c + depth] = district[r][c]
    return output


depth = 1
district = np.array([[2, 3, 4], [3, 4, 6], [2, 1, 4]])
print_matrix(district)
ans = create_fence(district, depth)
print_matrix(ans)
# This will print
# |  8  |  8  |  8  |  8  |  8  |
# -------------------------------
# |  8  |  2  |  3  |  4  |  8  |
# -------------------------------
# |  8  |  3  |  4  |  6  |  8  |
# -------------------------------
# |  8  |  2  |  1  |  4  |  8  |
# -------------------------------
# |  8  |  8  |  8  |  8  |  8  |
# -------------------------------
print("################")
depth = 2
district = np.array([[2, 3, 4, 1], [3, 4, 6, 5], [2, 1, 4, 7]])
print_matrix(district)
ans = create_fence(district, depth)
print_matrix(ans)
```

#Crows vs Cats


```python
def strength_difference(clubroom: np.ndarray) -> np.ndarray:
    rows, cols = clubroom.shape
    m = 2 * (rows - 1)
    strength_diff = np.zeros(m, dtype=int)
    count = 0
    for r in range(rows):
        for c in range(r + 1, cols):
            strength_diff[count] = clubroom[r][c] - clubroom[c][r]
            count += 1

    return strength_diff


clubroom = np.array([[1, 2, 9, 7], [4, 5, 1, 8], [3, 6, 2, 7], [2, 8, 6, 3]])
print_matrix(clubroom)
returned_value = strength_difference(clubroom)
print("Strength Difference Array is : ", returned_value)
unittest.output_test(returned_value, np.array([-2, 6, 5, -5, 0, 1]))
```

#ATM's Triangle


```python
def atm_triangle(n: int) -> np.ndarray:
    output = np.array([[None] * n] * n)
    rows = output.shape[0]
    for r in range(rows):
        output[r][0] = r + 1
        output[r][r] = r + 1
        if r >= 2:
            for i in range(r - 1):
                var = 0
                for j in range(i + 2):
                    var += output[r - 1][j]
                output[r][i + 1] = var
    return output


def print_without_none(matrix: np.ndarray) -> None:
    rows, cols = matrix.shape
    for r in range(rows):
        for c in range(cols):
            var = matrix[r][c]
            if var != None:
                print(var, end=" ")
        print()


n = 5
returned_value = atm_triangle(n)
print_without_none(returned_value)
# This should print
# 1
# 2  2
# 3  4  3
# 4  7  10  4
# 5  11 21  25  5
```

#Trace The BOT


```python
def make_move(current_pos: np.ndarray, cmd_value: int) -> np.ndarray:
    current_pos = np.array(current_pos)
    cmd_value -= 1
    match cmd_value // 4:
        case 0:
            current_pos[0] -= 2
            current_pos[1] -= 2
        case 1:
            current_pos[0] -= 2
            current_pos[1] += 2
        case 2:
            current_pos[0] += 2
            current_pos[1] -= 2
        case 3:
            current_pos[0] += 2
            current_pos[1] += 2

    match cmd_value % 4:
        case 0:
            current_pos[1] -= 1
        case 1:
            current_pos[1] += 1
        case 2:
            current_pos[0] -= 1
        case 3:
            current_pos[0] += 1

    return current_pos


def validate_cmd(current_pos: np.ndarray, cmd_value: int) -> np.ndarray:
    new_pos = make_move(current_pos, cmd_value)
    return new_pos if 0 <= new_pos[0] < 7 and 0 <= new_pos[1] < 7 else current_pos


def update_grid(grid: np.ndarray, current_pos: np.ndarray, value: str) -> None:
    row, col = current_pos
    grid[row][col] = value


def moving_around(commands: np.ndarray) -> np.ndarray:
    grid = np.full((7, 7), ".")
    current_pos = np.array([3, 3])
    update_grid(grid, current_pos, "-")
    n = len(commands)
    for i in range(n):
        cmd = commands[i]
        current_pos = validate_cmd(current_pos, cmd)
        update_grid(grid, current_pos, "*")
    update_grid(grid, current_pos, "/")
    return grid


cmds = np.array([5, 11, 2, 9])
result = moving_around(cmds)
print_matrix(result)
# This should print
# -------------------------------------------
# |  .  |  /  |  .  |  .  |  .  |  .  |  .  |
# -------------------------------------------
# |  .  |  .  |  .  |  .  |  *  |  .  |  .  |
# -------------------------------------------
# |  .  |  .  |  *  |  .  |  .  |  .  |  .  |
# -------------------------------------------
# |  .  |  .  |  .  |  -  |  .  |  .  |  .  |
# -------------------------------------------
# |  .  |  .  |  .  |  .  |  .  |  .  |  .  |
# -------------------------------------------
# |  .  |  .  |  .  |  .  |  .  |  .  |  .  |
# -------------------------------------------
# |  .  |  .  |  .  |  .  |  .  |  .  |  .  |
# -------------------------------------------
```

## MAIN CODE

```python
class Employee:
    # CONSTRUCTOR:
    def __init__(self, name:str, employee_id:int, dept:str, salary:int = 30000, employee_type:str = 'Regular') -> None:
        self.name:str = name
        self.employee_id:int = employee_id
        self.dept:str = dept
        self.salary:int = salary
        self.employee_type:str = employee_type
    
    # METHODS:
    def employeeDetails(self)->None:
        print(f"Name: {self.name}, Dept {self.dept}\nEmployee id: {self.employee_id}, Salary: {self.salary}")
        if self.employee_type != 'Regular':
            print(f"Employee Type: {self.employee_type}")
    
    def workDistribution(self, x_dept:str)->None:
        outputStr:str = 'Collect work distribution details from the manager.' if x_dept == 'Human Resource' else 'Collect work distribution loads from the HR department.'
        print(outputStr)
    
    def increment(self)->None:
        incr_dict:dict = {
            'Regular': 1.1,
            'Foreign': 1.15,
            'Part Time': 1
        }
        empType:str = self.employee_type
        self.salary*=incr_dict[empType]
        if empType == 'Part Time':
            print("Sadly, there is no increment for the part time employees!!")



class Foreign_employee(Employee):
    # CONSTRUCTOR
    def __init__(self, name: str, employee_id: int, dept: str, salary: int = 30000) -> None:
        super().__init__(name, employee_id, dept, 30000, 'Foreign')

class Part_time_employee(Employee):
    # CONSTRUCTOR
    def __init__(self, name: str, employee_id: int, dept: str, salary: int = 30000) -> None:
        super().__init__(name, employee_id, dept, salary, 'Part Time')
```

```python
class Point:
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y
        self.area = 0
    
    def calculate_area(self):
        return self.area
    
    def print_details(self):
        print("--------- Printing details ----------")
        print(f'Co-ordinate: ({self.x},{self.y})')
        print(f'Area: {self.area}')

class Circle(Point):
    def __init__(self, r, x=0, y=0):
        super().__init__(x, y)
        self.r = r
        self.area = 3.1416*r**2
    
    def print_details(self):
        super().print_details()
        print(f"Radius: {self.r}")

class Sphere(Circle):
    def __init__(self, r, x=0, y=0):
        super().__init__(r, x, y)
        self.area *= 4
```

## DRIVER CODE

```python
print("--------------1---------------")
p1 = Point(2,3)
print(f'Area of p1: {p1.calculate_area()}')
print("--------------2---------------")
p1.print_details()
print("--------------3---------------")
p2 = Point()
p2.print_details()
print("--------------4---------------")
c1 = Circle(4,0,3)
print(f'Area of c1: {c1.calculate_area()}')
print("--------------5---------------")
c1.print_details()
print("--------------6---------------")
c2 = Circle(7)
print(f'Area of c2: {c2.calculate_area()}')
print("--------------7---------------")
sph1 = Sphere(3,0,2)
print(f'Area of sph1: {sph1.calculate_area()}')
print("--------------8---------------")
sph1.print_details()
print("--------------9---------------")
sph2 = Sphere(6)
print(f'Area of sph2: {sph2.calculate_area()}')
```

## MAIN CODE

```python
class Account:
    def __init__(self, account_number, balance):
        self.account_number = account_number
        self.balance = balance
        self.account_type = "General"
        self.maturity = 0
    
    def print_details(self):
        print("------ Account details ------")
        print(f"Account Type: {self.account_type}, Maturity: {self.maturity} years")
        print(f"Account Number: {self.account_number}, Balance: ${self.balance:.2f}")
    
    def deposit(self, amount):
        self.balance += amount
        print(f"Deposited ${amount:.2f}. New Balance: ${self.balance:.2f}")
    
    def withdraw(self, amount):
        if self.balance >= amount:
            self.balance -= amount
            print(f"Withdrew ${amount:.2f}. New Balance: ${self.balance:.2f}")
        else:
            print("Insufficient funds.")
    
    def year_passed(self, year):
        self.maturity += year
        print(f"Maturity of the account: {self.maturity} years")

class SavingsAccount(Account):
    def __init__(self, account_type, account_number, balance, interest_rate, minimum_limit):
        super().__init__(account_number, balance)
        self.account_type = account_type
        self.interest_rate = interest_rate
        self.minimum_limit = minimum_limit

    def print_details(self):
        super().print_details()
        print(f"Interest Rate: {self.interest_rate}, Minimum Limit: ${self.minimum_limit}")
    def withdraw(self, amount):
        new_bal = self.balance - amount
        if new_bal>self.minimum_limit:
            super().withdraw(amount)
        else:
            print('Insufficient funds.')
    
    def apply_interest(self):
        self.balance*= (1 + self.interest_rate)
        print(f"Interest applied. New Balance: ${self.balance}")
        

class FixedDepositAccount(Account):
    def __init__(self,account_type, account_number, balance, min_year):
        super().__init__(account_number, balance)
        self.account_type = account_type
        self.min_year = min_year
        
    def deposit(self, amount):
        print("You can not deposit in a fixed deposit account.")

    def withdraw(self, amount):
        diff = self.min_year-self.maturity
        years_to_mature = max(0, diff)
        if years_to_mature:
            print(f"Account not yet matured. Wait {diff} year(s).")
        else:
            super().withdraw(amount)
        
```

## DRIVER CODE

```python
print("-----------1------------")
account = Account("A203", 2000)
account.print_details()
print("-----------2------------")
account.deposit(400)
account.withdraw(1500)
account.year_passed(2)
print("-----------3------------")
account.print_details()
print("-----------4------------")
savings_account = SavingsAccount("Savings","SA123", 1000, 0.05, 500)
savings_account.print_details()
print("-----------5------------")
savings_account.deposit(400)
print("-----------6------------")
savings_account.withdraw(1000)
print("-----------7------------")
savings_account.withdraw(800)
print("-----------8------------")
savings_account.apply_interest()
print("-----------9------------")
savings_account.print_details()
print("-----------10------------")
fixed_account1= FixedDepositAccount("Fixed Deposit","FDA321", 10000, 5)
fixed_account1.print_details()
print("-----------11------------")
fixed_account1.deposit(400)
print("-----------12------------")
fixed_account1.year_passed(6)
print("-----------13------------")
fixed_account1.withdraw(10000)
print("-----------14------------")
fixed_account1.print_details()
print("-----------15------------")
fixed_account2 = FixedDepositAccount("Fixed Deposit","FDA300", 50000, 7)
fixed_account2.print_details()
print("-----------16------------")
fixed_account2.withdraw(10000)
```

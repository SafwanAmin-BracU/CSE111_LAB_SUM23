```python
# Task 01
class Customer:
    def __init__(self):
        print("Welcome to ABC Memorial Park")
        self.tickets = []
    def buyTicket(self, name, age):
        if len(self.tickets) == 3:
            print("You can't buy more than 3 tickets")
        else:
            print(f"Successfully purchased a ticket for {name}!")
            price = 100 if age>10 else 50
            var = {'name': name, 'price': price}
            self.tickets.append(var)
    def showDetails(self):
        ticketCount = len(self.tickets)
        totalPrice = 0
        for ticket in self.tickets:
            totalPrice += ticket['price']
        print(f"Amount of tickets: {ticketCount}\nTotal price: {totalPrice} Taka")
        
# driver code
print('1-------------------------')
customer1 = Customer()
print('2-------------------------')
customer1.buyTicket('Bob', 23)
customer1.buyTicket('Henry', 7)
customer1.buyTicket('Alexa', 30)
customer1.buyTicket('Jonas', 43)
print('3-------------------------')
customer1.showDetails()
print('4-------------------------')
customer2 = Customer()
print('5-------------------------')
customer2.buyTicket('Harry', 60)
customer2.buyTicket('Tomas', 28)
print('6-------------------------')
customer2.showDetails()
```

```python
# Task 02
class SentenceAnalyzer:
    def __init__(self, sentence=""):
        self.sentence = sentence

    def set_sentence(self, sentence):
        self.sentence = sentence

    def get_word_count(self, length=0):
        if length == 0:
            print(f"Number of words in the sentence: {len(self.sentence.split())}")
        else:
            count = 0
            for word in self.sentence.split():
                if len(word) == length:
                    count += 1
            print(f"Number of words with {length} characters in the sentence: {count}")


# driver code
analyzer1 = SentenceAnalyzer()
analyzer1.set_sentence("That's an easy one")
print("1--------------------------")
analyzer1.get_word_count()
print("2--------------------------")
analyzer2 = SentenceAnalyzer("Like I said it's easy")
print("3--------------------------")
analyzer2.get_word_count()
print("4--------------------------")
analyzer2.get_word_count(4)
print("5--------------------------")
analyzer1.get_word_count(5)
```

```python
# Task 03
class Student:
    def __init__(self, name, cgpa, creditsCompleted=9, dept="CSE"):
        self.name = name
        self.cgpa = cgpa
        self.creditsCompleted = creditsCompleted
        self.enoughCredits = self.creditsCompleted > 10
        self.dept = dept
        SCHOLARSHIPS = {
            self.cgpa >= 3.7 and self.enoughCredits: "Merit-based",
            self.cgpa >= 3.5 and self.cgpa < 3.7 and self.enoughCredits: "Need-based",
            self.cgpa < 3.5 : "No",
        }
        self.scholarshipStatus = [x for x in SCHOLARSHIPS.values() if x][0]
        
        
    def checkScholarshipEligibility(self):
        if self.scholarshipStatus == "No":
            print(f"{self.name} is not eligible for any scholarship.")
        else:
            print(f"{self.name} is eligible for {self.scholarshipStatus} scholarship.")
    
    def showDetails(self):
        print(f"Name: {self.name}\nDepartment: {self.dept}\nCGPA: {self.cgpa}\nNumber of Credits: {self.creditsCompleted}\nScholarship Status: {self.scholarshipStatus}")

# driver code
print('--------------------------')
std1 = Student("Alif", 3.99, 12)
print('--------------------------')
std1.checkScholarshipEligibility()
print('--------------------------')
std1.showDetails()
print('--------------------------')
std2 = Student("Mim", 3.4)
std3 = Student("Henry", 3.5, 15,"BBA")
print('--------------------------')
std2.checkScholarshipEligibility()
print('--------------------------')
std3.checkScholarshipEligibility()
print('--------------------------')
std2.showDetails()
print('--------------------------')
std3.showDetails()
print('--------------------------')
std4 = Student("Bob", 4, 6, "CSE")
print('--------------------------')
std4.checkScholarshipEligibility()
print('--------------------------')
std4.showDetails()
```

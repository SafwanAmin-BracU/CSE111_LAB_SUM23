```python
# Task 03
class Farmer:
    def __init__(self, id=None) -> None:
        self.crops = []
        self.fishes = []
        ids = {
            type(id) == str: f"Welcome to your farm, {id}!",
            type(id) == int: f"Welcome to your farm. Your farm ID is {id}!"
        }
        if type(id) in [str, int]:
            print(ids[True])
        else:
            print("Welcome to your farm!")

    def addCrops(self, *crops):
        no_of_crops = len(crops)
        if no_of_crops:
            self.crops.extend(crops)
            print(f"{no_of_crops} crop(s) added.")
        else:
            print("No crop(s) added.")

    def addFishes(self, *fishes):
        no_of_fishes = len(fishes)
        if no_of_fishes:
            self.fishes.extend(fishes)
            print(f"{no_of_fishes} fish(s) added.")
        else:
            print("No fish added.")

    def showGoods(self):
        if self.crops:
            print(f"You have {len(self.crops)} crop(s):")
            all_crops = ""
            for crop in self.crops:
                all_crops += crop + ","
            print(all_crops[:-1])
        else:
            print("You don't have any crop(s).")
        if self.fishes:
            print(f"You have {len(self.fishes)} fish(s):")
            all_fishes = ""
            for fish in self.fishes:
                all_fishes += fish + ","
            print(all_fishes[:-1])
        else:
            print("You don't have any fish(s).")


f1 = Farmer()
print("-------------------")
f1.addCrops('Rice', "Jute", "Cinnamon")
print("-------------------")
f1.addFishes()
print("-------------------")
f1.addCrops('Mustard')
print("-------------------")
f1.showGoods()
print("-------------------")
f2 = Farmer("Korim Mia")
print("-------------------")
f2.addFishes('Pangash', 'Magur')
print("-------------------")
f2.addCrops("Wheat", "Potato")
print("-------------------")
f2.addFishes("Koi", "Tuna", "Sardine")
print("-------------------")
f2.showGoods()
print("-------------------")
f3 = Farmer(2865127000)
print("-------------------")
f3.addCrops()
print("-------------------")
f3.addFishes("Katla")
print("-------------------")
f3.showGoods()
print("-------------------")
```

```python
# Task 04
class UniversalStudiosUser:
    def __init__(self, name, age, type) -> None:
        self.name = name
        self.age = age
        self.type = type
        self.discount = self.type == "Special"
        self.rides = {}
        print("Welcome to Universal Studios.")

    def selected_rides(self, *rides):
        print("Added ride(s) successfully.")
        for ride in rides:
            ride_name, ride_price = ride.split("-")
            ride_price = int(ride_price)
            if ride_name in self.rides:
                self.rides[ride_name] += ride_price
            else:
                self.rides[ride_name] = ride_price

    def show_details(self):
        print(f"Your information:\nName: {self.name}, Age: {self.age}, Category: {self.type}\nSelected rides:")
        total_price = 0.0
        self.discount = self.discount and len(self.rides) > 3
        for ride, price in self.rides.items():
            print(f"Ride: {ride}, Amount: {price} dollar(s)")
            total_price += price
        if self.discount:
            print("Congratulations!!! You've got a 20% discount.")
            total_price *= 0.8
        print(f"Please pay {total_price} dollar(s).")


customer_1 = UniversalStudiosUser("Alice", 21,"Special")
print("--------- 1 ---------")
customer_1.selected_rides("Waterworld-100","Accelerator-200", "DinoSoarin-50")
print("--------- 2 ---------")
customer_1.show_details()
print("=================")
customer_2 = UniversalStudiosUser("Bob", 20,"Normal")
print("--------- 3 ---------")
customer_2.selected_rides("Enchanted Airways-300","Jurassic Park-500", "Accelerator-200","DinoSoarin-50")
print("--------- 4 ---------")
customer_2.show_details()
print("=================")
customer_3 = UniversalStudiosUser("Mark", 15,"Special")
print("--------- 5 ---------")
customer_3.selected_rides("Transformers-450","Jurassic Park-500", "Waterworld-100","DinoSoarin-50")
print("--------- 6 ---------")
customer_3.show_details()
```

```python
# Task 05

class Department:
    def __init__(self, dept_name='ChE', no_of_sections=5) -> None:
        self.dept_name = dept_name
        self.no_of_sections = no_of_sections
        self.students = []
        self.average = 0
        print(
            f"The {self.dept_name} Department has {self.no_of_sections} sections.")

    def add_students(self, *students):
        self.students.extend(students)
        if len(students) != self.no_of_sections:
            print(
                f"The {self.dept_name} Department doesn't have {len(students)} sections.")
        else:
            total_students = 0
            for section in students:
                total_students += section
            self.average = total_students / self.no_of_sections
            print(
                f"The {self.dept_name} Department has an average of {self.average} students in each section.")

    def merge_Department(self, *departments):
        total_students = len(self.students)*self.average
        for dept in departments:
            print(f"{dept.dept_name} Department is merged to {self.dept_name}.")
            var = len(dept.students)*dept.average
            total_students += var
        self.average = total_students / self.no_of_sections
        return f"The Engineering Department has an average of {self.average} students in each section."


d1 = Department()
print('1-----------------------------------')
d2 = Department('MME Department')
print('2-----------------------------------')
d3 = Department('NCE Department', 8)
print('3-----------------------------------')
d1.add_students(12, 23, 12, 34, 21)
print('4-----------------------------------')
d2.add_students(40, 30, 21)
print('5-----------------------------------')
d3.add_students(12, 34, 41, 17, 30, 22, 32, 51)
print('6-----------------------------------')
mega = Department('Engineering Department', 10)
print('7-----------------------------------')
mega.add_students(21, 30, 40, 36, 10, 32, 27, 51, 45, 15)
print('8-----------------------------------')
print(mega.merge_Department(d1, d2))
print('9-----------------------------------')
print(mega.merge_Department(d3))
```

```python
# Task 06
class StudentRoutineGenerator:
    def __init__(self, name, courses) -> None:
        self.name = name
        self.no_of_course = courses
        self.routine = {
            'Sat/Thurs': {},
            'Sun/Tue': {},
            'Mon/Wed': {}
        }
        self.courses = {}
        print(
            f"Name: {self.name}\nMaximum Number of Courses: {self.no_of_course}\nInitial Routine: {self.routine}")

    def showRoutine(self):
        print(f"Updated Routine:\n{self.routine}\nRoutine Details:")
        for day in self.routine:
            if not self.routine[day]:
                continue
            classes = self.routine[day]
            print(f"{day}:")
            for time, course in classes.items():
                print(f"{time} - {course}")

    def addCourses(self, *courses):
        for course in courses:
            if len(self.courses) == self.no_of_course:
                print(f"You can't take more than {self.no_of_course} courses")
                break
            course_name, class_date, class_time = course.split('-')
            if course_name in self.courses:
                print(f"You already have {course_name} in your routine")
            else:
                if class_time in self.routine[class_date] and self.routine[class_date][class_time] != course_name:
                    print(
                        f"Can't take {course_name}. It's clashing with your {self.routine[class_date][class_time]}")
                else:
                    print(f"Successfully added {course_name}!")
                    self.routine[class_date][class_time] = course_name
                    self.courses[course_name]= (class_date, class_time)

    def dropCourse(self, course):
        if course not in self.courses:
            print(f"No such course in your routine")
        else:
            class_date, class_time = self.courses[course]
            del self.routine[class_date][class_time]
            print(f"Successfully dropped {course}!")

print('########################################')
st1 = StudentRoutineGenerator('Harry', 4)
print('------------------------------')
st1.addCourses('CSE110-Mon/Wed-12:30', 'MAT110-Mon/Wed-2:00')
st1.addCourses('ENG101-Sun/Tue-12:30', 'CSE110-Mon/Wed-9:30')
st1.addCourses('PHY111-Sun/Tue-12:30')
print('------------------------------')
st1.showRoutine()
print('------------------------------')
st1.dropCourse('CSE110')
st1.dropCourse('PHY112')
print('------------------------------')
st1.showRoutine()
print('########################################')
st2 = StudentRoutineGenerator('John', 3)
print('------------------------------')
st2.addCourses('MAT110-Mon/Wed-8:00')
st2.addCourses('ENG101-Sat/Thurs-12:30', 'CSE110-Sun/Tue-9:30')
st2.addCourses('PHY111-Sun/Tue-12:30')
print('------------------------------')
st2.showRoutine()
```

```python
class KK_tea:
    total_sales_dict: dict = {}

    def __init__(self, price: int, qty: int = 50, name: str = 'Regular') -> None:
        self.name: str = f'KK {name} Tea'
        self.price: int = price
        self.qty: int = qty
        self.weight: int = self.qty*2
        self.status: bool = False
        KK_tea.total_sales_dict[self.name] = 0

    def product_detail(self) -> None:
        print(
            f'Name: {self.name}, Weight: {self.weight}\nTea Bags: {self.qty}, Price: {self.price}\nStatus: {self.status}')

    @classmethod
    def total_sales(cls) -> None:
        print(f'Total sales: {cls.total_sales_dict}')

    @classmethod
    def update_sold_status_regular(cls, *objs: object) -> None:
        for obj in objs:
            obj.status = True
            cls.total_sales_dict[obj.name] += 1


class KK_flavoured_tea(KK_tea):
    def __init__(self, name: str, price: int, qty: int = 50) -> None:
        super().__init__(price, qty, name)

    @classmethod
    def update_sold_status_flavoured(cls, *objs: object) -> None:
        KK_tea.update_sold_status_regular(*objs)
```

#### DRIVER CODE

```python
t1 = KK_tea(250)
print("-----------------1-----------------")
t1.product_detail()
print("-----------------2-----------------")
KK_tea.total_sales()
print("-----------------3-----------------")
t2 = KK_tea(470, 100)
t3 = KK_tea(360, 75)
KK_tea.update_sold_status_regular(t1, t2, t3)
print("-----------------4-----------------")
t3.product_detail()
print("-----------------5-----------------")
KK_tea.total_sales()
print("-----------------6-----------------")
t4 = KK_flavoured_tea("Jasmine", 260, 50)
t5 = KK_flavoured_tea("Honey Lemon", 270, 45)
t6 = KK_flavoured_tea("Honey Lemon", 270, 45)
print("-----------------7-----------------")
t4.product_detail()
print("-----------------8-----------------")
t6.product_detail()
print("-----------------9-----------------")
KK_flavoured_tea.update_sold_status_flavoured(t4, t5, t6)
print("-----------------10-----------------")
KK_tea.total_sales()
```

```python
class TwoDVector:
    def __init__(self, x: int, y: int) -> None:
        self.__x: int = x
        self.__y: int = y
    # GETTERS:
    def get_x(self) -> int: return self.__x
    def get_y(self) -> int: return self.__y

    # SETTERS:
    def set_x(self, n_x: int) -> None:
        self.__x = n_x

    def set_y(self, n_y: int) -> None:
        self.__y = n_y

    # METHODS:
    def add2DVectors(self, *vectors):
        for vector in vectors:
            self.__x += vector.get_x()
            self.__y += vector.get_y()

    def print2DVector(self, parent: bool = False):
        y = "+ "+str(self.__y) if self.__y >= 0 else str(self.__y)
        endstr = '' if parent else '\n'
        print(f"{self.__x}i {y}j", end=endstr)


class ThreeDVector(TwoDVector):
    def __init__(self, x: int, y: int, z: int) -> None:
        super().__init__(x, y)
        self.__z: int = z

    # GETTERS
    def get_z(self) -> int: return self.__z

    # SETTERS:
    def set_z(self, n_z: int) -> None:
        self.__z = n_z

    # METHODS
    def add3DVectors(self, *vectors) -> None:
        self.add2DVectors(*vectors)
        for vector in vectors:
            self.__z += vector.get_z()

    def print3DVector(self) -> None:
        self.print2DVector(True)
        z = "+ "+str(self.__z) if self.__z >= 0 else str(self.__z)
        print(f' {z}k')

    def multiplyScalar(self, n: int) -> None:
        self.set_x(self.get_x()*n)
        self.set_y(self.get_y()*n)
        self.set_z(self.get_z()*n)

    def calculateLength(self) -> float:
        return (self.get_x()**2+self.get_y()**2+self.get_z()**2)**0.5
```

#### DRIVER CODE

```python
TwoDV1 = TwoDVector(5, 6)
TwoDV2 = TwoDVector(3, 7)
TwoDV3 = TwoDVector(1, 8)
print("===============")
TwoDV1.add2DVectors(TwoDV2, TwoDV3)
TwoDV1.print2DVector()
print("===============")
ThreeDV1 = ThreeDVector(5, 6, 1)
ThreeDV2 = ThreeDVector(1, 9, -7)
ThreeDV3 = ThreeDVector(8, 2, 4)
print("===============")
ThreeDV1.add3DVectors(ThreeDV2,ThreeDV3)
ThreeDV1.print3DVector()
print("===============")
ThreeDV1.multiplyScalar(3)
ThreeDV1.print3DVector()
print("===============")
print(ThreeDV1.calculateLength())
```

```python
class File:

    files_directory: list[object] = []

    def __init__(self, name: str, size: str, date_of_creation: str):

        self.name: str = name
        self.size: str = size
        self.date_of_creation: str = date_of_creation
        self.description: str = ''
        File.files_directory.append(self)

    def renameFile(self, n):
        self.name = n

    def updateDescription(self, txt):
        self.description = txt

    def previewFile(self):
        print(f"Name of file: {self.name}")
        print(f"Description: {self.description}")
        print(f"Size: {self.size}")
        print(f"Date created: {self.date_of_creation}")

    @classmethod
    def previewAllFiles(cls):
        for obj in cls.files_directory:
            print('===')
            obj.previewFile()


class Audio(File):
    # Write your code here
    def __init__(self, name: str, size: str, date_of_creation: str, duration, lyrics: dict = {}):
        super().__init__(name, size, date_of_creation)
        self.duration: str = duration
        self.lyrics: dict = lyrics

    def playAudio(self) -> None:
        lyrics = self.lyrics
        print(f'Playing Audio file now for {self.duration}', end='')
        if lyrics:
            print(': (Lyric is given below):\n')
            for k, v in lyrics.items():
                print(f'{k} : {v}')
        else:
            print(' (no lyric available).')


class Image(File):
    # Write your code here
    def __init__(self, name: str, size: str, date_of_creation: str, dimension: str):
        super().__init__(name, size, date_of_creation)
        dimension = dimension.split('x')
        self.width: int = int(dimension[0])
        self.height: int = int(dimension[1])

    def displayPicture(self) -> None:
        picture = [self.width*'*']*self.height
        print(*picture, sep='\n')
```

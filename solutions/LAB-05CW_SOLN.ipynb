### Task 01

Please Write the **Student** and **Department** class with necessary properties so that the provided driver code generates output given below. Make sure the **ID** and **CGPA** attributes in the **'Student'** class are private and cannot be accessed directly from outside the class


```python
# Main Code:

class Student:
    # CONSTRUCTOR:
    def __init__(self, name: str, id: int, cgpa: float) -> None:
        self.name = name
        self.__id = id
        self.__cgpa = cgpa

    # GETTERS AND SETTERS:
    def getId(self) -> int:
        return self.__id

    def getcgpa(self) -> float:
        return self.__cgpa

    def setId(self, id: int) -> None:
        self.__id = id

    def setcgpa(self, cgpa: float) -> None:
        self.__cgpa = cgpa

    # DUNDERS:
    def __repr__(self) -> str:
        return f"Student Name: {self.name}, ID: {self.__id}, cgpa: {self.__cgpa}"


class Department:
    # CONSTRUCTOR:
    def __init__(self, name: str) -> None:
        self.name: str = name
        self.students: list[Student] = []
        self.all_ids: list[int] = []

    # METHODS:

    def addStudent(self, *students: list[Student]) -> None:
        for student in students:
            if student.getId() not in self.all_ids:
                self.students.append(student)
                self.all_ids.append(student.getId())
                print(f"Welcome to CSE department, {student.name}")
            else:
                print(
                    f"Student with the same ID {student.getId()} already exists, Please try with another ID")

    def findStudent(self, id: int) -> Student:
        for student in self.students:
            if student.getId() == id:
                return student
        return None

    def details(self):
        print(f"Department Name: {self.name}")
        for student in self.students:
            print(student)
```

```python
# Driver Code:

s1 = Student("Akib", 22301010, 3.29)
s2 = Student("Reza", 22101010, 3.45)
s3 = Student("Ruhan", 23101934, 4.00)
print("1==================================")
cse = Department("CSE")
cse.findStudent(22112233)
print("2==================================")
cse.addStudent(s1, s2, s3)
print("3==================================")
cse.details()
print("4==================================")
cse.findStudent(22301010)
print("5==================================")
s4 = Student("Nakib", 22301010, 3.22)
cse.addStudent(s4)
print("6==================================")
s4.setId(21201220)
cse.addStudent(s4)
print("7==================================")
cse.details()
print("8==================================")
s5 = Student("Sakib", 22201010, 2.29)
cse.addStudent(s5)
print("9==================================")
cse.details()
```

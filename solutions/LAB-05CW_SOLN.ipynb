### Task 01

Please Write the **Student** and **Department** class with necessary properties so that the provided driver code generates output given below. Make sure the **ID** and **CGPA** attributes in the **'Student'** class are private and cannot be accessed directly from outside the class

````python
# Main Code:

class Student:
    # CONSTRUCTOR:
    def __init__(self, x_name: str, x_id: int, x_cgpa: float) -> None:
        self.name = x_name
        self.__id = x_id
        self.__cgpa = x_cgpa

    # GETTERS AND SETTERS:
    def getId(self) -> int:
        return self.__id

    def getcgpa(self) -> float:
        return self.__cgpa

    def setId(self, n_id: int) -> None:
        self.__id = n_id

    def setcgpa(self, n_cgpa: float) -> None:
        self.__cgpa = n_cgpa

    # DUNDERS:
    def __repr__(self) -> str:
        return f"Student Name: {self.name}, ID: {self.__id}, cgpa: {self.__cgpa}"


class Department:

    # CONSTRUCTOR:
    def __init__(self, name: str) -> None:
        self.name: str = name
        self.students: dict[int, Student] = {}

    # METHODS:
    def addStudent(self, *students) -> None:

        # LOOP THROUGH STUDENTS:
        for student in students:

            # GET STUDENT DETAILS:
            x_id = student.getId()
            x_name = student.name
            
            # CHECK IF STUDENT ALREADY EXISTS:
            if x_id not in self.students.keys():
                self.students[x_id] = student
                print(f"Welcome to {self.name} department, {x_name}")
            else:
                print("Student with the same ID already exists, Please try with another ID")

    def findStudent(self, x_id: int) -> Student:
        
        # GET STUDENT (None if not found):
        student = self.students.get(x_id)

        # IF STUDENT EXISTS:
        if student:
            print(f"Student info:\nStudent Name: {student.name}\nID: {x_id}\nCGPA: {student.getcgpa()}")
        else: 
            print(f"Student with this ID doesn't exist, Please give a valid ID")
    
    
    def details(self):
        print(f"Department Name: {self.name}\nNumber of student:{len(self.students)}")
        for student in self.students.values():
            print(student)
````

````python
# Driver Code:

s1 = Student("Akib", 22301010, 3.29)
s2 = Student("Reza", 22101010, 3.45)
s3 = Student("Ruhan", 23101934, 4.00)
print("1==================================")
cse = Department("CSE")
cse.findStudent(22112233)
print("2==================================")
cse.addStudent(s1, s2, s3)
print("3==================================")
cse.details()
print("4==================================")
cse.findStudent(22301010)
print("5==================================")
s4 = Student("Nakib", 22301010, 3.22)
cse.addStudent(s4)
print("6==================================")
s4.setId(21201220)
cse.addStudent(s4)
print("7==================================")
cse.details()
print("8==================================")
s5 = Student("Sakib", 22201010, 2.29)
cse.addStudent(s5)
print("9==================================")
cse.details()
````

### Task 02

##### Class Description:
* Spaceship: This class represents a spaceship. Each spaceship has a name and a capacity (the maximum weight it can carry).

* Cargo: This class represents a piece of cargo. Each cargo item has a name and a weight. Both attributes should be private which means they cannot be accessed directly from outside of the class.

A Spaceship contains (HAS) Cargo. That means each spaceship can carry multiple cargo items, but the total weight of the cargo cannot exceed the spaceship's capacity. 

````python
# Main Code:

class Cargo:
    # CONSTRUCTOR:
    def __init__(self, x_name: str, x_weight: int) -> None:
        self.__name = x_name
        self.__weight = x_weight

    # GETTERS AND SETTERS:
    def get_name(self) -> str:
        return self.__name

    def get_weight(self) -> int:
        return self.__weight

    def set_name(self, x_name: str) -> None:
        self.__name = x_name

    def set_weight(self, x_weight: int) -> None:
        self.__weight = x_weight

    # DUNDERS:
    def __repr__(self) -> str:
        return f"{self.__name}"


class Spaceship:
    # CONSTRUCTOR:
    def __init__(self, x_name: str, x_capacity: int) -> None:
        self.name = x_name
        self.capacity = x_capacity
        self.cargo: list[Cargo] = []
        self.weight: int = 0
    
    # METHODS:
    def load_cargo(self, cargo: Cargo) -> None:
        
        # Get cargo attributes:
        x_name = cargo.get_name()
        x_weight = cargo.get_weight()
                
        # Check if there is space:
        has_space = self.weight + x_weight <= self.capacity
        
        # Load cargo if there is space:
        if has_space:
            self.cargo.append(cargo)
            self.weight += x_weight
        else:
            exceed = self.weight + x_weight - self.capacity
            print(f"Warning: Unable to load {x_name} inside {self.name}. Exceeds capacity by {exceed}.")
    
    def display_details(self) -> None:
        print(f"Spaceship Name: {self.name}\nCapacity: {self.capacity}\nCurrent Cargo Weight: {self.weight}\nCargo: {self.cargo}")
````

````python
# Driver Code:

# Creating spaceships
falcon = Spaceship("Falcon", 50000)
apollo = Spaceship("Apollo", 100000)
enterprise = Spaceship("Enterprise", 220000)
print("1.===================================")
# Creating cargo
gold = Cargo("Gold", 20000)
platinum = Cargo("Platinum", 25000)
dilithium = Cargo("Dilithium", 50000)
trilithium = Cargo("Trilithium", 70000)
neutronium = Cargo("Neutronium", 80000)
print("2.===================================")
# Loading cargo onto spaceships
falcon.load_cargo(gold)
falcon.load_cargo(platinum)
falcon.display_details()
print("3.===================================")
apollo.load_cargo(gold) # Apollo will not reach its total capacity
apollo.display_details()
print("4.===================================")
falcon.load_cargo(neutronium) # This should exceed Falcon's capacity
print("5.===================================")
enterprise.load_cargo(dilithium)
enterprise.load_cargo(trilithium)
enterprise.load_cargo(neutronium) # This should not exceed Enterprise's capacity
enterprise.display_details()
````

### Task 03

For this task, you will be designing classes for a music player app. There are 2 classes involved:

##### Song Class

The `Song` class has:

- 3 private instance variables:
  - `title` (private): A string containing the song title.
  - `artist` (private): A string containing the artist title.
  - `duration` (private): An integer representing the song duration time in seconds.

- An overloaded parameterized constructor to initialize the instance variables.
  - Parameters for `title` and `duration` are required, but `artist` is optional.
  - Default value for `artist`: "Unknown Artist".

- `info()`: A method to return a string formatted as:
  - Example: "{title} by {artist} ({min}:{sec})"
  - One example: "Never Gonna Give You Up by Rick Astley (3:33)"

##### Playlist Class

The `Playlist` class has:

- 3 private instance variables:
  - `name` (private): A name for the playlist.
  - `song_list` (private): A list containing `Song` objects.
  - `now_id` (private): An integer representing the current song index in `song_list`.

- An overloaded parameterized constructor to initialize the instance variables.
  - Parameters for `name` are required, `song_list` is optional, and `now_id` is always set by default.
  - Default value for `song_list`: An empty list, and `now_id`: -1.

- `play_next()`: A method that updates `now_id` to the next index. If this method is called while the last song is being played, it will change the `now_id` variable to the first index of the `song_list`.

- `play_prev()`: Updates `now_id` to the previous index. If this method is called while the first song in the playlist is being played, it will show an error message like this: "Cannot go back from the first song.", then play the first song again.

- `add_songs(*songs)`: A method that adds all `Song` objects from the parameter `songs` to its `song_list`.

- `remove_song(song_obj)`: A method to remove/delete `song_obj` from `song_list`.

- `show_playlist()`: A method that prints some playlist information.
  - An example output would be:
    ```
    Playlist: Classical Orchestra
    Total Songs: 3
    1. The Flower Duet by Leo Delibes (1:26)
    2. Winter by Vivaldi (3:43)
    3. Ride of the Valkyries by Wagner (2:17)
    ```

- Appropriate getter and setter methods for accessing the variables.

- When displaying song details, `play_next()`, `play_prev()`, and `show_playlist()` methods must call the `info()` method for each `Song` object.

Your tasks are:

- Implement the `Song` and `Playlist` classes ensuring proper encapsulation. All data members in both classes should be private and have corresponding getter and setter methods.

- Implement the necessary methods described above.


````python
# Main Code:

class Song:
    # CONSTRUCTOR:
    def __init__(self, *args) -> None:
        # Check if the user has provided the title and duration:
        if len(args) < 2:
            print('Cannot create song without title and duration!')
            title = '--- Error ---'
            duration = 0
            artist = '--- Error ---'
        else:
            title, *rest, duration = args
            artist = rest[0] if rest else 'Unknown Artist'

        # Set the attributes:
        self.__title: str = title
        self.__duration: int = duration
        self.__artist: str = artist

    # GETTERS AND SETTERS:

    # Getters:
    def get_title(self) -> str:
        return self.__title

    def get_duration(self) -> int:
        return self.__duration

    def get_artist(self) -> str:
        return self.__artist

    # Setters:
    def set_title(self, n_title) -> None:
        self.__title = n_title

    def set_duration(self, n_duration) -> None:
        self.__duration = n_duration

    def set_artist(self, n_artist) -> None:
        self.__artist = n_artist

    # METHODS:

    def info(self) -> str:
        mins = self.__duration // 60
        secs = self.__duration % 60
        return f'{self.__title} by {self.__artist} ({mins}:{secs})'


class Playlist:
    # CONSTRUCTOR:
    def __init__(self, x_name: str, *x_song_list) -> None:
        self.__name: str = x_name
        self.__song_list: list[Song] = [*x_song_list]
        self.__now_idx: int = -1

    # GETTERS AND SETTERS:

    # Getters:
    def get_name(self) -> str:
        return self.__name

    def get_song_list(self) -> list[Song]:
        return self.__song_list

    def get_now_idx(self) -> int:
        return self.__now_idx

    # Setters:
    def set_name(self, n_name: str) -> None:
        self.__name = n_name

    def set_song_list(self, n_song_list: list[Song]) -> None:
        self.__song_list = n_song_list

    def set_now_idx(self, n_now_idx: int) -> None:
        self.__now_idx = n_now_idx

    # METHODS:
    def play_song(self) -> None:
        song: Song = self.__song_list[self.__now_idx]
        print(f'Now playing: {song.info()}')

    def play_next(self) -> None:
        self.__now_idx += 1
        if self.__now_idx == len(self.__song_list):
            print("Reached the end of the playlist, starting from the first song.")
            self.__now_idx = 0
        self.play_song()

    def play_prev(self) -> None:
        if self.__now_idx == 0:
            print("Cannot go back from the first song.")
            self.play_song()
        else:
            self.__now_idx -= 1
            self.play_song()

    def add_songs(self, *songs) -> None:
        print(f'Added {len(songs)} song(s) to the playlist {self.__name}')
        self.__song_list.extend(songs)

    def remove_song(self, x_song: Song) -> None:
        if x_song in self.__song_list:
            self.__song_list.remove(x_song)
            print(f'Removed {x_song.get_title()} from {self.__name}.')
        else:
            print(f'{x_song.get_title()} is not in {self.__name}.')

    def show_playlist(self) -> None:
        print(
            f"Playlist: {self.__name}\nTotal Songs: {len(self.__song_list)}")
        for idx, song in enumerate(self.__song_list):
            print(f'{idx+1}. {song.info()}')
````
